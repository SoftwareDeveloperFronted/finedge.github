# Data Preprocessing

```python
df.isnull().sum()
```

```python
# handle numerical missing data

df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
df.isnull().sum()
```

```python
# handle categorical missing data

df['Gender'].mode()[0]
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

# Train Test split

```python
x=df.iloc[:,:-1]
x.head(3)
```

```python
y=df["Loan_Status (Approved)_Y"]
```

```python
x.shape
```

```python
y.shape
```

```python
# For train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=25) # here for random state we can take from 1 to 1000
```

```python
# x_train.head()
x_train.head(3)
```

```python
y_train
```

```python
x_test
```

# Model Training

```python
# After spliting data we made model training 
# we use Random forest classifier and train to model

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

model.fit(x_train, y_train)
```

```python
# print accuracy of this model

print("Accuracy is", model.score(x_test, y_test)*100)
```

```python
# We take decision tree classifier

from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(x_train, y_train)
print("Accuracy is", model2.score(x_test,y_test)*100)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
model3.fit(x_train, y_train)
print("Accuracy is", model3.score(x_test, y_test)*100)
```

```python
# Confusion matrics

from sklearn.metrics import confusion_matrix
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```
